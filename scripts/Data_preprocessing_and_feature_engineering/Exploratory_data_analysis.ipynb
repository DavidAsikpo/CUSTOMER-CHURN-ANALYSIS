```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.core.pylabtools import figsize
from narwhals import read_csv

```

```python
df = pd.read_csv('/Users/mac/PycharmProjects/Customer_Churn/Merged_Dataset.csv')
df.head(10)
```

```python
for column_names in df.columns:
    df[column_names] = df[column_names].fillna(0)


```

```python
df
```

```python
plt.figure(figsize=(8,4))
plt.scatter(df.index, df['total_amount_spent'])
plt.title("Scatter Plot of Amount Spent")
plt.xlabel("Index")
plt.ylabel("Amount Spent")
plt.show()
```

```python
Q1 = df['total_amount_spent'].quantile(0.25)
Q3 = df['total_amount_spent'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['total_amount_spent'] < lower_bound) | (df['total_amount_spent'] > upper_bound)]
print("Number of outliers:", len(outliers))
```

```python
Q1 = df['average_spent'].quantile(0.25)
Q3 = df['average_spent'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['average_spent'] < lower_bound) | (df['average_spent'] > upper_bound)]
print("Number of outliers:", len(outliers))
```

```python
plt.scatter(x = df['number_of_transactions'], y = df['total_amount_spent'], c=df['average_spent'], cmap='Accent')
plt.title("Scatter Plot of Total Amount Spent/Number of Transactions")
plt.colorbar(label="Average Spent per Transaction")
plt.xlabel("Number of Transactions")
plt.ylabel("Total Amount Spent")
plt.show()
```

```python
columns_interest = df.columns
correlation_matrix = df[columns_interest].drop('service_usage', axis = 1).corr(method="spearman", )
plt.figure(figsize = (14,8))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu",fmt= ".2f", cbar_kws = {"label": "Spearman Correlation"})
heatmap.set_title('Correlation Heatmap')
plt.show()
```

```python
corr_matrix = df.drop('service_usage', axis = 1).corr().abs()
```

```python
threshold = 0.8
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
high_corr_features = [column for column in upper.columns if any(upper[column] > threshold)]
print("Highly correlated features:", high_corr_features)
```
